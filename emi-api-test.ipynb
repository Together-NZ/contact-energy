```python
import urllib.request, json

try:
    url = "https://emi.azure-api.net/real-time-dispatch/"
    hdr = {
        'Cache-Control': 'no-cache',
        'Ocp-Apim-Subscription-Key': '148c8e2b4c1b4f95a88440dc3be43c39',
    }
    req = urllib.request.Request(url, headers=hdr)
    req.get_method = lambda: 'GET'
    response = urllib.request.urlopen(req)
    
    raw_data = response.read()
    data = json.loads(raw_data.decode('utf-8'))
    
    # Filter for a specific PointOfConnectionCode
    target_poc = "ABY0111"
    for entry in data:
        if entry["PointOfConnectionCode"] == target_poc:
            print(f"PoC: {entry['PointOfConnectionCode']}")
            print(f"Time: {entry['FiveMinuteIntervalDatetime']}")
            print(f"Load (MW): {entry['SPDLoadMegawatt']}")
            print(f"Generation (MW): {entry['SPDGenerationMegawatt']}")
            print(f"Price ($/MWh): {entry['DollarsPerMegawattHour']}")
            print()

except Exception as e:
    print("Error:", e)
```

```python
total_price = sum(entry["DollarsPerMegawattHour"] for entry in data)
avg_price = total_price / len(data)
print(f"Total Price: ${total_price:.2f}, Average Price ($/MWh): {avg_price:.2f}")
```

```python
generators = [entry for entry in data if entry["SPDGenerationMegawatt"] > 0]
loads = [entry for entry in data if entry["SPDLoadMegawatt"] != 0]

print(f"Number of Generators: {len(generators)}")
for gen in generators[:5]:  # First 5 for brevity
        print(f"Generator {gen['PointOfConnectionCode']}: {gen['SPDGenerationMegawatt']} MW, ${gen['DollarsPerMegawattHour']}/MWh")

print(f"\nNumber of Loads: {len(loads)}")
for load in loads[:5]:
        print(f"Load {load['PointOfConnectionCode']}: {load['SPDLoadMegawatt']} MW, ${load['DollarsPerMegawattHour']}/MWh")
```

```python
max_price_entry = max(data, key=lambda x: x["DollarsPerMegawattHour"])
min_price_entry = min(data, key=lambda x: x["DollarsPerMegawattHour"])
    
print(f"Highest Price: ${max_price_entry['DollarsPerMegawattHour']} at {max_price_entry['PointOfConnectionCode']}")
print(f"Lowest Price: ${min_price_entry['DollarsPerMegawattHour']} at {min_price_entry['PointOfConnectionCode']}")
```
